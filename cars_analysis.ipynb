### 1. Find all Null Values in the dataset. If there is any null value in any column, than fill it with the Mean of the column.

```python
# null areas
df.isnull().sum()
```

```python
mean = df['Cylinders'].mean()

df['Cylinders'].fillna(mean, inplace=True)
```

### 2. Based on value counts

### 3. Instructions (Filtering)
#### . Show all the records where origin is Asia or Europe

```python
df.head()
```

```python
# First way
df[(df['Origin'] == 'Asia') | (df['Origin'] == 'Europe')]
```

```python
# Second way
df[df['Origin'].isin(['Asia','Europe'])]
```

### 4. Instructions (Removing unwanted records)
#### . Remove all the records (rows) where Weight is above 4000

```python
# ~ will remove all the records specified after it

df[~df['Weight'] > 4000]

```

### 5. Instruction (Apply function as a column)
#### . Increase all the values of 'MGP_City' column by 3.

```python
# Way 1
df['MPG_City'] = ( df['MPG_City'] + 3)
```

```python
# way 2
df['MPG_City'] = df['MPG_City'].apply(lambda x:x+3)
```
